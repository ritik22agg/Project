```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Model,Sequential
from keras.applications.resnet50 import ResNet50
import os
from keras.utils import np_utils
from keras.layers import Dense, Dropout
```

```python
model = ResNet50(include_top = False, weights = 'imagenet', pooling = 'avg')
model.summary()
```

```python
fc1 = Dense(1024,activation='sigmoid')(model.output)
dp = Dropout(0.3)(fc1)
fc2 = Dense(512, activation= 'sigmoid')(dp)
dp2 = Dropout(0.4)(fc2)
fc3 = Dense(64)(dp2)
dp3 = Dropout(0.2)(fc3)
fc4 = Dense(2, activation='softmax')(dp3)

model_new = Model(inputs = model.input, outputs = fc4)
model_new.summary()
```

```python
#data preprocessing 
X_data = []
labels = []
label_2_index = {'Deepika' : 1, 'Ritik' : 0}
index_2_label = {0:'Ritik', 1:'Deepika'}
for file in os.listdir('/home/ritik/Desktop/Image_classificationNeuralNetwork/Image_data'):
    x = np.load('/home/ritik/Desktop/Image_classificationNeuralNetwork/Image_data/' + file)
    label = file.split('.')[0]
    for ix in range(x.shape[0]):
        X_data.append(x[ix])
        labels.append(label_2_index[label])
    print(len(labels))
```

```python
X_data = np.array(X_data)
labels = np.array(labels)
print(X_data.shape, labels.shape)
```

```python
labels = np_utils.to_categorical(labels)
print(labels.shape)
print(labels.shape)
```

```python
print(labels)
```

```python
model_new.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
for ix in range(len(model_new.layers)):
    print(ix,model_new.layers[ix])
```

```python
for ix in range(176):
    model_new.layers[ix].trainable = False

model_new.summary()
```

```python
model_new.fit(X_data, labels,shuffle = True, batch_size=2,epochs=5)
```

```python
model_new.save_weights('/home/ritik/Desktop/Image_classificationNeuralNetwork/model_weights.h5')
```
